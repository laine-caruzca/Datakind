## Clustering

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 8) 
```

```python
main_data['LSOA Name'][]
```

```python
#1) standardise data

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(main_data.dropna())
```

```python
main_data.index = main_data['LSOA Name']
main_data=main_data.drop(['LSOA Name','LSOA Code'], axis = 'columns')
```

```python
tmp = main_data.isna().sum()
tmp[tmp>0]
```

```python
main_data = main_data.drop(['Children aged 0-4 in poverty','Children aged 11-15 in poverty','Children aged 16-19 in poverty','Children aged 5-10 in poverty'], axis='columns')
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=15)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.plot(pca.explained_variance_ratio_)
```

```python
import numpy as np
plt.plot(np.cumsum(pca.explained_variance_ratio_))
```

```python
results = []
```

```python
for i in range(1,30):
    kmeans = KMeans(n_clusters=i, random_state=0).fit(X_pca)
    results.append(kmeans.inertia_)
```

```python
plt.plot(results)
```

```python
kmeans = KMeans(n_clusters=10, random_state=0).fit(X_pca)
```

```python
labels = kmeans.predict(X_pca)
```

```python
len(labels)
```

```python
len(main_data.dropna())
```

```python
cleandata = main_data.dropna()
```

```python
cleandata['clusters'] = labels
```

```python
cleandata.head()
```

```python
cluster_property = cleandata.groupby('clusters').mean()
```

```python
cluster_property.transpose()
```

```python
merged_data = merged_data.merge(cleandata[['clusters']],left_on='LSOA Name',right_index=True)
```

```python
ax = plt.figure(figsize=(10,10)).gca()
merged_data.plot(column='clusters', cmap='tab20b', legend = True, ax=ax)
```
